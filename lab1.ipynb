```python
plt.figure(figsize = (10, 6))

plt.scatter(data[data['Gender'] == 'Male']['Weight'],
            data[data['Gender'] == 'Male']['Height'],
            alpha = 0.15,
            label = 'Male',
            color = 'b')

plt.scatter(data[data['Gender'] == 'Female']['Weight'],
            data[data['Gender'] == 'Female']['Height'],
            alpha = 0.05,
            label = 'Female',
            color = 'r')

plt.xlabel('Weight')
plt.xticks(range(0,200,10))
plt.xlim(0, 199)
plt.ylabel('Height')
plt.legend()
plt.grid()
```
